```python
# Define bins and labels
import pandas as pd
bins = [0, 5, 10, 15, 20, 25, 30, 35]  # 7 bin edges, thus 6 bins
labels = ['0-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35']  # 7 labels for 7 bins

# Convert 'PTS' to numeric if needed (not necessary here since 'PTS' is already numeric)
sdf['PTS'] = pd.to_numeric(sdf['PTS'], errors='coerce')

# Cut 'PTS' into categorical bins
sdf['PTS'] = pd.cut(sdf['PTS'], bins=bins, labels=labels, right=False, ordered=True)

print("\nDataFrame with PTS as Categorical:")
print(sdf)
```

```python
pts_dummies = pd.get_dummies(sdf['PTS'], prefix='PTS').astype(int)
sdf = pd.concat([sdf, pts_dummies], axis=1)

print("\nDataFrame after Binarization of PTS:")
print(sdf)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sdf = pd.DataFrame({
    'PTS': [15, 20, 10, 25, 30],
    'AST': [5, 10, 8, 15, 20],
    'REB': [10, 8, 12, 7, 5],
    'STL': [2, 4, 3, 5, 1],
    'BLK': [1, 3, 2, 4, 0]
})

# Compute the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Add labels and title
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
